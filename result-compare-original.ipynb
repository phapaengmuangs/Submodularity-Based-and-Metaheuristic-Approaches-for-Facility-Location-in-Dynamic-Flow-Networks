```python
#For read CSV data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
result_gd_m = pd.read_csv("result-bd-gd-momentum.csv")
result_gd_s = pd.read_csv("result-bd-gd-step.csv")
result_nm = pd.read_csv("result-bd-nelder-mead.csv")
```

```python
gd_m_flow_t = result_gd_m['flow time'].tolist()
gd_s_flow_t = result_gd_s['flow time'].tolist()
nm_flow_t = result_nm['flow time'].tolist()

gd_m_cost_t = result_gd_m['cost time'].tolist()
gd_s_cost_t = result_gd_s['cost time'].tolist()
nm_cost_t = result_nm['cost time'].tolist()

gd_m_time = []
for i in range(len(gd_m_flow_t)):
    gd_m_time.append(gd_m_flow_t[i]+gd_m_cost_t[i])

gd_s_time = []
for i in range(len(gd_s_flow_t)):
    gd_s_time.append(gd_s_flow_t[i]+gd_s_cost_t[i])

nm_time = []
for i in range(len(nm_flow_t)):
    nm_time.append(nm_flow_t[i]+nm_cost_t[i])

k = np.arange(len(gd_m_time))

plt.plot(k, nm_time, ':', linewidth = 2, color='#ffa600', label='Nelder-Mead')
plt.plot(k, gd_m_time, '--', linewidth = 2, color='#0000ff', label='Momentum')
plt.plot(k, gd_s_time, '-.', linewidth = 2, color='#00b300', label='Step')

plt.xlabel("Number of facilities")
plt.ylabel("Time")
plt.title("Comparing time used for solving Problem 1")

plt.legend()
plt.savefig('compare-time-big.pdf')
plt.show()
```

```python
gd_m_flow = result_gd_m['flow(t)'].tolist()
gd_s_flow = result_gd_s['flow(t)'].tolist()
nm_flow = result_nm['flow(t)'].tolist()

k = np.arange(len(gd_m_flow))

plt.plot(k, nm_flow, ':', linewidth = 2, color='#ffa600', label='Nelder-Mead')
plt.plot(k, gd_m_flow, '--', linewidth = 2, color='#0000ff', label='Momentum')
plt.plot(k, gd_s_flow, '-.', linewidth = 2, color='#00b300', label='Step')

plt.xlabel("Number of facilities")
plt.ylabel("Number of Evacuees")
plt.title("Comparing result of number of evacuees on Problem 1 and 1A")

plt.legend()
plt.savefig('compare-flow-big.pdf')
plt.show()
```

```python
iter_gd_m = pd.read_csv("iteration-bd-gd-momentum.csv")
iter_gd_s = pd.read_csv("iteration-bd-gd-step.csv")
iter_nm = pd.read_csv("iteration-bd-nelder-mead.csv")
```

```python
i = 4

while i <= 9:
    gd_m_iter = iter_gd_m[str(i)].tolist()
    gd_s_iter = iter_gd_s[str(i)].tolist()
    nm_iter = iter_nm[str(i)].tolist()

    k1 = np.arange(len(nm_iter))
    k2 = np.arange(len(gd_m_iter))
    k3 = np.arange(len(gd_s_iter))

    plt.plot(k1, nm_iter, ':', linewidth = 2, color='#ffa600', label='Nelder-Mead')
    plt.plot(k2, gd_m_iter, '--', linewidth = 2, color='#0000ff', label='Momentum')
    plt.plot(k3, gd_s_iter, '-.', linewidth = 2, color='#00b300', label='Step')

    n = i+1
    plt.xlabel("Iterations")
    plt.ylabel("cost(x)")
    plt.title("Comparing value of cost(x) with "+ str(n) + ' facilities')

    plt.legend()
    plt.savefig('compare-cost-'+str(n)+'-big.pdf')
    plt.show()
    i+=1
```
